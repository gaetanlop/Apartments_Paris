```python
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium import webdriver
import time
import pandas as pd
url='https://www.logic-immo.com/appartement-paris/vente-appartement-paris-75-100_1-1.html'
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path="OneDrive\chromedriver.exe", options=options)
driver.set_window_size(1120, 1000)

df_house=[]
for page in range(40,250):
    driver.get(f'https://www.logic-immo.com/appartement-paris/vente-appartement-paris-75-100_1-{page}.html')
    houses=driver.find_elements_by_xpath('.//a[@class="offer-link"]')

    for i in range(len(houses)-1):
        
        try:
            superficie=driver.find_elements_by_xpath('.//span[@class="offer-details-caracteristik--area"]')[i].text
        except:
            superficie=-1
        try:
            price=driver.find_elements_by_xpath('.//p[@class="offer-price"]')[i].text
        except:
            price=-1    
        try:
            location=driver.find_elements_by_xpath('.//div[@class="offer-details-location"]')[i].text     
        except:
            location=-1 
        try:
            bedrooms=driver.find_elements_by_xpath('.//div[@class="offer-details-caracteristik"]')[i].find_element_by_xpath('.//span[@class="offer-details-caracteristik--bedrooms"]').text
        except:
            bedrooms=-1
        try:
            rooms=driver.find_elements_by_xpath('.//span[@class="offer-details-caracteristik--rooms"]')[i].text
        except:
            rooms=-1
     

        try:    
            test=driver.find_elements_by_xpath('.//a[@class="offer-link"]')[i]
            driver.execute_script("arguments[0].click();", test)
            time.sleep(2)
        except:
            driver.get(f'https://www.logic-immo.com/appartement-paris/vente-appartement-paris-75-100_1-{page}.html')
            continue
            
        try:    
            driver.window_handles[1]
            driver.switch_to.window(driver.window_handles[1])
            driver.close()
            driver.switch_to.window(driver.window_handles[0])   
            continue
        except:
            try:
                desc=driver.find_element_by_xpath('.//ul[@class="list"]').text
            except:
                desc=-1

        driver.execute_script("window.history.go(-1)")

        df_house.append({"Superficie" : superficie,
                "Rooms" : rooms,
                "Bedrooms" : bedrooms,
                "Price" : price,
                "Location" : location,
                "description":desc})
    print(page)
```

```python
df=pd.DataFrame(df_house)
df
```

```python
df.to_csv('house_logic_immo1.csv',index=False)
```
